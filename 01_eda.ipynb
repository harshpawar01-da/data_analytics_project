```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset
df = pd.read_csv("../data/processed/mumbai_air_quality_clean.csv")

# Convert date column to datetime
if "date" in df.columns:
    df["date"] = pd.to_datetime(df["date"], errors="coerce")

# Quick overview
print(df.head())
print(df.info())
print(df.describe())

# Plot distribution of PM2.5
plt.figure(figsize=(10,6))
sns.histplot(df["pm25"].dropna(), bins=50, kde=True)   # <-- fixed column name
plt.title("Distribution of PM2.5 Levels in Mumbai")
plt.xlabel("PM2.5")
plt.ylabel("Frequency")
plt.show()

# Trend over time (monthly average)
if "date" in df.columns:
    df.set_index("date", inplace=True)
    df["pm25"].resample("M").mean().plot(figsize=(12,6))   # <-- fixed column name
    plt.title("Monthly Average PM2.5 Levels in Mumbai")
    plt.ylabel("PM2.5")
    plt.xlabel("Date")
    plt.show()
```
